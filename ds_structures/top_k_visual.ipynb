```python
import heapq

def top_k_visual(nums, k):
    heap = []
    for n in nums:
        heapq.heappush(heap, n)
        print("Heap:", heap)

        if len(heap) > k:
            removed = heapq.heappop(heap)
            print("Removed:", removed)

    print("Top K elements:", heap)


if __name__ == "__main__":
    top_k_visual([5, 1, 9, 2, 7, 3], 3)
```
