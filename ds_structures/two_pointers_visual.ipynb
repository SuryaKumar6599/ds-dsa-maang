```python
def two_pointers_visual(nums, target):
    left, right = 0, len(nums) - 1

    while left < right:
        s = nums[left] + nums[right]
        print(f"Checking {nums[left]} + {nums[right]} = {s}")

        if s == target:
            print(f"✅ Found pair: {nums[left]}, {nums[right]}")
            return
        elif s < target:
            left += 1
        else:
            right -= 1

    print("❌ No pair found")


if __name__ == "__main__":
    two_pointers_visual([1, 2, 3, 4, 6], 6)
```
