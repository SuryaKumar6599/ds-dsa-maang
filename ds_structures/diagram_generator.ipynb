```python
def two_heaps_diagram(low, high):
    print("\nLOW (Max Heap)     HIGH (Min Heap)")
    print("-" * 35)

    max_len = max(len(low), len(high))
    for i in range(max_len):
        l = str(-low[i]) if i < len(low) else ""
        h = str(high[i]) if i < len(high) else ""
        print(f"{l:<15} {h}")


if __name__ == "__main__":
    low = [-3, -2, -1]
    high = [4, 5]
    two_heaps_diagram(low, high)
```
